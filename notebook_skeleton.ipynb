```python
import os
import matplotlib.pyplot as plt
import numpy as np

import package as pkg

rootpath = os.getcwd()
path_to_rirs = os.path.join(rootpath, "rirs")
path_to_soundfiles = os.path.join(rootpath, "sound_files")

sound_filenames = [
    os.path.join(path_to_soundfiles, "speech1.wav"),
    os.path.join(path_to_soundfiles, "speech2.wav"),
]
noise_filenames = [os.path.join(path_to_soundfiles, "White_noise1.wav")]
signal_duration = 10  # [s]
```

Build acoustic scenario and generate accompanying RIRs using GUI

```python
## launch GUI
pkg.RIRg_GUI(
    exportFolder=path_to_rirs,
    roomDim=10,  # [m]
    distBwMics=10,  # [cm]
    fs=44100,  # [Hz]
    t60=0,  # [s]
)
```

Load the latest acoustic scenario and plot some of the data that's in there.
